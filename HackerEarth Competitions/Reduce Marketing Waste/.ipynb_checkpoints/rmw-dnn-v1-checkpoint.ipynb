## Import libraries

```python
import gc
import pickle
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import StratifiedKFold

from tensorflow.keras.regularizers import l2
from tensorflow.keras.utils import plot_model
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow_addons.optimizers import AdamW, Lookahead
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.layers import Embedding, Reshape
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.layers import BatchNormalization, Add
from tensorflow.keras.layers import Concatenate, Activation
```

## Prepare data for model training

```python
with open("../input/rmw-data-preprocess-v2/Reduce_Marketing_Waste_Dataset.txt", 'rb') as handle: 
    data = handle.read()

processed_data = pickle.loads(data)
train_df = processed_data['train_df']
test_df = processed_data['test_df']

del processed_data
gc.collect()
```

```python
cat_cols = ["Pitch","Lead_revenue","Fund_category","Lead_source",
            "create_dt_year","create_dt_day_weekend",
            "Last_lead_update","Resource","Internal_rating"]

for col in cat_cols:
    dummy_val = pd.get_dummies(train_df[col], prefix=col)
    train_df = pd.concat([train_df, dummy_val], axis=1)
    train_df.drop([col], inplace=True, axis=1)

print("train_df: {}".format(train_df.shape))

for col in cat_cols:
    dummy_val = pd.get_dummies(test_df[col], prefix=col)
    test_df = pd.concat([test_df, dummy_val], axis=1)
    test_df.drop([col], inplace=True, axis=1)

print("test_df: {}".format(test_df.shape))
```

```python
cat_cols = ["Lead_name_type","Industry","Hiring_candidate_role","Internal_POC"]

train_df[cat_cols] = train_df[cat_cols].astype(int)
test_df[cat_cols] = test_df[cat_cols].astype(int)

vocab_size1 = train_df['Lead_name_type'].nunique()
vocab_size2 = train_df['Industry'].nunique()
vocab_size3 = train_df['Hiring_candidate_role'].nunique()
vocab_size4 = train_df['Internal_POC'].nunique()

vocab_list = [vocab_size1, vocab_size2, vocab_size3, vocab_size4]
```

```python
Xtest1 = np.array([test_df['Lead_name_type'].values]).T
Xtest2 = np.array([test_df['Industry'].values]).T
Xtest3 = np.array([test_df['Hiring_candidate_role'].values]).T
Xtest4 = np.array([test_df['Internal_POC'].values]).T

test_df.drop(["Lead_name_type","Industry","Hiring_candidate_role","Internal_POC"], 
             axis=1, inplace=True)

Xtest5 = test_df.values
print("Xtest5: {}".format(Xtest5.shape))

test_list = [Xtest1, Xtest2, Xtest3, Xtest4, Xtest5]

del test_df
gc.collect()
```

## Build and validate the model

```python
def dnn_model(vocab_list, n_features):

    x_input1 = Input(shape=(1, ), name='Input-Lead_name_type')
    x_input2 = Input(shape=(1, ), name='Input-Industry')
    x_input3 = Input(shape=(1, ), name='Input-Hiring_candidate_role')
    x_input4 = Input(shape=(1, ), name='Input-Internal_POC')
    x_input5 = Input(shape=(n_features, ), name='Input-continuous_features')
    
    embed_dim = int(min(np.ceil(vocab_list[0]/2), 15))
    x1 = Embedding(input_dim=int(vocab_list[0])+1, output_dim=embed_dim, 
                   name="Embed-Lead_name_type")(x_input1)
    x1 = Reshape(target_shape=(embed_dim,))(x1)
    
    embed_dim = int(min(np.ceil(vocab_list[1]/2), 35))
    x2 = Embedding(input_dim=int(vocab_list[1])+1, output_dim=embed_dim, 
                   name="Embed-Industry")(x_input2)
    x2 = Reshape(target_shape=(embed_dim,))(x2)
    
    embed_dim = int(min(np.ceil(vocab_list[2]/2), 45))
    x3 = Embedding(input_dim=int(vocab_list[2])+1, output_dim=embed_dim, 
                   name="Embed-Hiring_candidate_role")(x_input3)
    x3 = Reshape(target_shape=(embed_dim,))(x3)
    
    embed_dim = int(min(np.ceil(vocab_list[3]/2), 15))
    x4 = Embedding(input_dim=int(vocab_list[3])+1, output_dim=embed_dim, 
                   name="Embed-Internal_POC")(x_input4)
    x4 = Reshape(target_shape=(embed_dim,))(x4)
    
    x5 = Dense(units=40, kernel_initializer='he_uniform', 
               kernel_regularizer=l2(0.0001))(x_input5)
    x5 = BatchNormalization()(x5)
    
    x = Concatenate(name='Concatenate-all-inputs')([x1, x2, x3, x4, x5])
    x = BatchNormalization()(x)

    x = Dense(units=64, kernel_initializer='he_uniform', 
                kernel_regularizer=l2(0.0001))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(rate=0.5)(x)
    
    x = Dense(units=64, kernel_initializer='he_uniform', 
                kernel_regularizer=l2(0.0001))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(rate=0.5)(x)
    
    x_output = Dense(units=1, kernel_initializer='he_uniform')(x)

    model = Model(inputs=[x_input1, x_input2, x_input3, x_input4, x_input5], 
                  outputs=x_output, 
                  name='DNN_Model')
    return model
```

```python
model = dnn_model(vocab_list, Xtest5.shape[1])
model.compile(loss='mse', 
              metrics=[RootMeanSquaredError(name='rmse')],
              optimizer=Lookahead(AdamW(lr=1e-2, 
                                        weight_decay=1e-5, 
                                        clipvalue=700), 
                                  sync_period=10))
```

```python
plot_model(
    model, to_file='./DNN_model.png', 
    show_shapes=True, show_layer_names=True
)
```

```python
FOLD = 5
NUM_SEED = 1
VERBOSE = 0

np.random.seed(3)
seeds = np.random.randint(0, 100, size=NUM_SEED)
Ytrain_strat = pd.qcut(train_df['Success_probability'].values, q=5, labels=range(0,5))

oof_score = 0
y_pred_meta_dnn = np.zeros((train_df.shape[0], NUM_SEED))
y_pred_final_dnn = np.zeros((Xtest5.shape[0], NUM_SEED))
counter = 0
mini_batch_size = 128


for sidx, seed in enumerate(seeds):
    seed_score = 0
    
    kfold = StratifiedKFold(n_splits=FOLD, shuffle=True, random_state=seed)

    for idx, (train, val) in enumerate(kfold.split(train_df.loc[:, train_df.columns!='Success_probability'], Ytrain_strat)):
        counter += 1

        train_df1, val_df = train_df.iloc[train].copy(), train_df.iloc[val].copy()
        
        Xtrain1, Xval1 = np.array([train_df1['Lead_name_type'].values]).T, np.array([val_df['Lead_name_type'].values]).T
        Xtrain2, Xval2 = np.array([train_df1['Industry'].values]).T, np.array([val_df['Industry'].values]).T
        Xtrain3, Xval3 = np.array([train_df1['Hiring_candidate_role'].values]).T, np.array([val_df['Hiring_candidate_role'].values]).T
        Xtrain4, Xval4 = np.array([train_df1['Internal_POC'].values]).T, np.array([val_df['Internal_POC'].values]).T
        
        Ytrain = np.log1p(np.abs(np.array([train_df1['Success_probability'].values]).T))
        Yval = np.log1p(np.abs(np.array([val_df['Success_probability'].values]).T))

        train_df1.drop(["Lead_name_type","Industry","Hiring_candidate_role",
                        "Internal_POC","Success_probability"], axis=1, inplace=True)
        val_df.drop(["Lead_name_type","Industry","Hiring_candidate_role",
                     "Internal_POC","Success_probability"], axis=1, inplace=True)

        Xtrain5, Xval5 = train_df1.values, val_df.values
        
        train_list = [Xtrain1, Xtrain2, Xtrain3, Xtrain4, Xtrain5]
        val_list = [Xval1, Xval2, Xval3, Xval4, Xval5]

        model = dnn_model(vocab_list, Xtest5.shape[1])
        model.compile(loss='mse', 
                      metrics=[RootMeanSquaredError(name='rmse')],
                      optimizer=Lookahead(AdamW(lr=5e-3, 
                                                weight_decay=1e-5, 
                                                clipvalue=700), 
                                          sync_period=10))

        early = EarlyStopping(monitor="val_rmse", mode="min", 
                              restore_best_weights=True, 
                              patience=10, verbose=VERBOSE)

        reduce_lr = ReduceLROnPlateau(monitor="val_rmse", factor=0.2, 
                                      min_lr=1e-6, patience=5, 
                                      verbose=VERBOSE, mode='min')

        chk_point = ModelCheckpoint('./DNN_model.h5', 
                                    monitor='val_rmse', verbose=VERBOSE, 
                                    save_best_only=True, mode='min')
        
        history = model.fit(
            train_list, Ytrain, 
            batch_size=mini_batch_size,
            epochs=150, 
            verbose=VERBOSE, 
            workers=5,
            callbacks=[reduce_lr, early, chk_point], 
            validation_data=(val_list, Yval)
        )
        
        model = load_model('./DNN_model.h5')

        y_pred = model.predict(val_list)
        y_pred_meta_dnn[val, sidx] += y_pred.ravel()
        y_pred_final_dnn[:, sidx] += model.predict(test_list).ravel()
        
        score = np.sqrt(mean_squared_error(Yval, y_pred))
        oof_score += score
        seed_score += score
        print("Seed-{} | Fold-{} | OOF Score: {}".format(seed, idx, score))
    
    print("\nSeed: {} | Aggregate OOF Score: {}\n\n".format(seed, (seed_score / FOLD)))


y_pred_final_dnn = y_pred_final_dnn / float(FOLD)
oof_score /= float(counter)
print("Aggregate OOF Score: {}".format(oof_score))
```

```python
np.savez_compressed('./DNN_Meta_Features.npz',
                    y_pred_meta_dnn=y_pred_meta_dnn, 
                    oof_score=oof_score,
                    y_pred_final_dnn=y_pred_final_dnn)
```
