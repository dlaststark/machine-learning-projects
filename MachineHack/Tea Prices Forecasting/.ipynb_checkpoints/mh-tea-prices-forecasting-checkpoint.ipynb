## Load source datasets

```python
train_df = pd.read_csv("../input/mh-tea-story-dataset/train.csv", 
                       parse_dates=['WeekEnding_Date'], 
                       dayfirst=True)
print(f"train_df: {train_df.shape}")
train_df.head()
```

```python
test_df = pd.read_csv("../input/mh-tea-story-dataset/test.csv", 
                      parse_dates=['WeekEnding_Date'], 
                      dayfirst=True)
test_df = test_df.sort_values(by=['WeekEnding_Date']).copy()
print(f"test_df: {test_df.shape}")
test_df.head()
```

```python
missing_df = pd.read_csv("../input/mh-missing-data/Missing Data.csv", 
                         parse_dates=['WeekEnding_Date'], 
                         dayfirst=True)
missing_df.head()
```

## Data Cleansing

```python
train_df = pd.concat([train_df, missing_df])
train_df = train_df.sort_values(by=['WeekEnding_Date']).copy()
print(f"train_df: {train_df.shape}")
train_df.head()
```

```python
train_df.drop_duplicates(subset=['WeekEnding_Date'], keep='first', inplace=True)
train_df[train_df['WeekEnding_Date']==pd.Timestamp(datetime.date(2013, 3, 16))]

train_df = train_df[train_df['WeekEnding_Date']!=pd.Timestamp(datetime.date(2010, 10, 13))].copy()
train_df = train_df[train_df['WeekEnding_Date']!=pd.Timestamp(datetime.date(2018, 4, 4))].copy()
```

```python
for col in tqdm([x for x in test_df.columns if x != 'WeekEnding_Date']):
    train_df[col] = train_df[col].apply(lambda x: np.nan if x in ['N. S.','NS','N.S','N.S.','No Sale','No sale','N.S.\xa0'] else float(str(x).replace('NS','')))
    test_df[col] = test_df[col].apply(lambda x: np.nan if x in ['N. S.','NS','N.S','N.S.','No Sale','No sale','N.S.\xa0'] else float(str(x).replace('NS','')))
```

```python
train_df['Darjeeling_Ref_Price'] = train_df['Darjeeling_Ref_Price'].apply(lambda x: np.nan if x==0 else x)
train_df['Ernakulam_Average_Price'] = train_df['Ernakulam_Average_Price'].apply(lambda x: np.nan if x==0 else x)
train_df['Siliguri_Average_Price'] = train_df['Siliguri_Average_Price'].apply(lambda x: np.nan if x<40 else x)
train_df['Guwahati_Average_Price'] = train_df['Guwahati_Average_Price'].apply(lambda x: np.nan if x==0 else x)
```

## Handle missing values

```python
combined_df = train_df.append(test_df, sort=False, 
                              ignore_index=False)

combined_df['year'] = combined_df['WeekEnding_Date'].map(lambda x: pd.to_datetime(x).year)
combined_df['quarter'] = combined_df['WeekEnding_Date'].map(lambda x: pd.to_datetime(x).quarter)
combined_df['month'] = combined_df['WeekEnding_Date'].map(lambda x: pd.to_datetime(x).month)
combined_df['week'] = combined_df['WeekEnding_Date'].map(lambda x: pd.to_datetime(x).week) 
combined_df['day_year'] = combined_df['WeekEnding_Date'].map(lambda x: pd.to_datetime(x).dayofyear) 
combined_df['day_month'] = combined_df['WeekEnding_Date'].map(lambda x: pd.to_datetime(x).day) 
combined_df['season'] = combined_df['month'].apply(lambda x: 0 if x in [2,3] else 1 if x in [4,5,6] else 2 if x in [7,8] else 3 if x in [9,10,11] else 4)
combined_df['day_week'] = combined_df['WeekEnding_Date'].map(lambda x: pd.to_datetime(x).dayofweek) 
combined_df['day_weekend'] = np.where(combined_df['day_week'].isin([5,6]),1,0)
combined_df['date_diff_days'] = combined_df['WeekEnding_Date'].apply(lambda x: (pd.Timestamp(datetime.date(2020, 1, 1)) - x).days)
combined_df['date_diff_years'] = combined_df['date_diff_days'].apply(lambda x: round(x/365, 3))
combined_df['date_diff_months'] = combined_df['date_diff_days'].apply(lambda x: round(x/30, 3))
combined_df['is_holiday'] = combined_df.apply(lambda x: 1 if x['WeekEnding_Date'] in holidays.IN(years=x['year']) else 0, axis=1)
combined_df.head(10)
```

```python
col_list = [
    'Kolkata_Average_Price', 'Kolkata_Ref_Price',
    'Bangalore_Average_Price', 'Bangalore_Ref_Price',
    'Cochin_Average_Price', 'Cochin_Ref_Price',
    'Darjeeling_Average_Price', 'Darjeeling_Ref_Price',
    'Ernakulam_Average_Price', 'Ernakulam_Ref_Price',
    'Siliguri_Average_Price', 'Siliguri_Ref_Price',
    'Guwahati_Average_Price', 'Guwahati_Ref_Price'
]

for col in tqdm(col_list):
    combined_df[col] = combined_df[col].transform(lambda x: x.ffill())
    combined_df[col] = combined_df[col].transform(lambda x: x.bfill())

combined_df[combined_df.isin([np.nan, np.inf, -np.inf]).any(1)].head()
```

```python
seas_d = sm.tsa.seasonal_decompose(train_df['Average'], model='add', period=200);
fig = seas_d.plot()
fig.set_figheight(7)
fig.set_figwidth(12);
```

```python
def test_adf(series):
    dfout={}
    dftest=sm.tsa.adfuller(series.dropna(), autolag='AIC', regression='ct')
    
    for key,val in dftest[4].items():
        dfout[f'critical value ({key})']=val
    
    print(f"   P-value: {dftest[1]}")
    if dftest[1]<=0.055:
        print("   Reject Null Hypothesis - Data is Stationary")
        
    else:
        print("   Accept Null Hypothesis - Data is NOT Stationary")
        
    return dftest[1]
```

```python
col_list = [
    'Kolkata_Average_Price', 'Kolkata_Ref_Price',
    'Bangalore_Average_Price', 'Bangalore_Ref_Price',
    'Cochin_Average_Price', 'Cochin_Ref_Price',
    'Darjeeling_Average_Price', 'Darjeeling_Ref_Price',
    'Ernakulam_Average_Price', 'Ernakulam_Ref_Price',
    'Siliguri_Average_Price', 'Siliguri_Ref_Price',
    'Guwahati_Average_Price', 'Guwahati_Ref_Price',
    'Average'
]

for col in col_list:
    print(col)
    
    p = test_adf(train_df[col])
    if p>0.055:
        print("\n   Trying to make data stationary...")
        p = test_adf(train_df[col].diff())
        
    print("")
```

```python
def plot_corr(series):
    fig,ax = plt.subplots(2,1, figsize=(15, 6))
    fig = sm.tsa.graphics.plot_acf(series, lags=52, ax=ax[0])
    fig = sm.tsa.graphics.plot_pacf(series, lags=52, ax=ax[1])
    fig.tight_layout();
```

## Feature Engineering

```python
combined_df['Kolkata_Price_Diff'] = combined_df['Kolkata_Average_Price'] - combined_df['Kolkata_Ref_Price']
combined_df['Bangalore_Price_Diff'] = combined_df['Bangalore_Average_Price'] - combined_df['Bangalore_Ref_Price']
combined_df['Cochin_Price_Diff'] = combined_df['Cochin_Average_Price'] - combined_df['Cochin_Ref_Price']
combined_df['Darjeeling_Price_Diff'] = combined_df['Darjeeling_Average_Price'] - combined_df['Darjeeling_Ref_Price']
combined_df['Ernakulam_Price_Diff'] = combined_df['Ernakulam_Average_Price'] - combined_df['Ernakulam_Ref_Price']
combined_df['Siliguri_Price_Diff'] = combined_df['Siliguri_Average_Price'] - combined_df['Siliguri_Ref_Price']
combined_df['Guwahati_Price_Diff'] = combined_df['Guwahati_Average_Price'] - combined_df['Guwahati_Ref_Price']

combined_df['Kolkata_Price_Diff%'] = (np.abs(combined_df['Kolkata_Price_Diff'])*100)/combined_df['Kolkata_Ref_Price']
combined_df['Bangalore_Price_Diff%'] = (np.abs(combined_df['Bangalore_Price_Diff'])*100)/combined_df['Bangalore_Ref_Price']
combined_df['Cochin_Price_Diff%'] = (np.abs(combined_df['Cochin_Price_Diff'])*100)/combined_df['Cochin_Ref_Price']
combined_df['Darjeeling_Price_Diff%'] = (np.abs(combined_df['Darjeeling_Price_Diff'])*100)/combined_df['Darjeeling_Ref_Price']
combined_df['Ernakulam_Price_Diff%'] = (np.abs(combined_df['Ernakulam_Price_Diff'])*100)/combined_df['Ernakulam_Ref_Price']
combined_df['Siliguri_Price_Diff%'] = (np.abs(combined_df['Siliguri_Price_Diff'])*100)/combined_df['Siliguri_Ref_Price']
combined_df['Guwahati_Price_Diff%'] = (np.abs(combined_df['Guwahati_Price_Diff'])*100)/combined_df['Guwahati_Ref_Price']

combined_df['Average_Ref_Price'] = (combined_df['Kolkata_Ref_Price'] + combined_df['Bangalore_Ref_Price'] + \
                                    combined_df['Cochin_Ref_Price'] + combined_df['Darjeeling_Ref_Price'] + \
                                    combined_df['Ernakulam_Ref_Price'] + combined_df['Siliguri_Ref_Price'] + \
                                    combined_df['Guwahati_Ref_Price']) / 7.0

combined_df['Average_Sale_Price'] = (combined_df['Kolkata_Average_Price'] + combined_df['Bangalore_Average_Price'] + \
                                     combined_df['Cochin_Average_Price'] + combined_df['Darjeeling_Average_Price'] + \
                                     combined_df['Ernakulam_Average_Price'] + combined_df['Siliguri_Average_Price'] + \
                                     combined_df['Guwahati_Average_Price']) / 7.0

combined_df['Avg_Price_Diff'] = combined_df['Average_Sale_Price'] - combined_df['Average_Ref_Price']
combined_df['Avg_Price_Diff%'] = (np.abs(combined_df['Avg_Price_Diff'])*100)/combined_df['Average_Ref_Price']

combined_df.replace([np.nan, np.inf, -np.inf], 0, inplace=True)
combined_df.head()
```

### Lag Features

```python
window_dict = {
    'Kolkata_Average_Price': [1, 3, 4], 
    'Kolkata_Ref_Price': [1, 3, 4], 
    'Bangalore_Average_Price': [1, 4], 
    'Bangalore_Ref_Price': [1, 4], 
    'Cochin_Average_Price': [1, 2], 
    'Cochin_Ref_Price': [1, 2], 
    'Darjeeling_Average_Price': [1, 2], 
    'Darjeeling_Ref_Price': [1, 2],
    'Ernakulam_Average_Price': [1, 2, 3], 
    'Ernakulam_Ref_Price': [1, 2, 3],
    'Siliguri_Average_Price': [1, 2, 3], 
    'Siliguri_Ref_Price': [1, 2, 3],
    'Guwahati_Average_Price': [1, 2], 
    'Guwahati_Ref_Price': [1, 2]
}

for col in tqdm(window_dict.keys()):
    for w in window_dict.get(col):
        combined_df[col+'_Lag'+str(w)] = combined_df.groupby(['year'])[col].shift(w)
        combined_df[col+'_LagDiff'+str(w)] = np.abs(combined_df[col] - combined_df[col+'_Lag'+str(w)])

combined_df.fillna(-1, inplace=True)
combined_df.tail()
```

```python
window_dict = {
    'Kolkata_Price_Diff': [1, 2, 52], 
    'Kolkata_Price_Diff%': [1, 2, 51], 
    'Bangalore_Price_Diff': [1, 2, 3, 52], 
    'Bangalore_Price_Diff%': [1, 3], 
    'Cochin_Price_Diff': [1, 2, 52], 
    'Cochin_Price_Diff%': [1], 
    'Darjeeling_Price_Diff': [1, 2],
    'Darjeeling_Price_Diff%': [1], 
    'Ernakulam_Price_Diff': [1, 2, 3], 
    'Ernakulam_Price_Diff%': [1, 2, 3, 9, 10, 11],
    'Siliguri_Price_Diff': [1, 2, 3],
    'Siliguri_Price_Diff%': [1, 2, 3], 
    'Guwahati_Price_Diff': [1, 2], 
    'Guwahati_Price_Diff%': [1], 
    'Average_Ref_Price': [1, 2],
    'Average_Sale_Price': [1, 2],
    'Avg_Price_Diff': [1], 
    'Avg_Price_Diff%': [1, 2],
    'Average': [1, 2, 14, 52]
}

for col in tqdm(window_dict.keys()):
    for w in window_dict.get(col):
        combined_df[col+'_Lag'+str(w)] = combined_df[col].shift(w)
        combined_df[col+'_LagDiff'+str(w)] = np.abs(combined_df[col] - combined_df[col+'_Lag'+str(w)])

combined_df.fillna(-1, inplace=True)
combined_df.tail()
```

## Final predictions

```python
test_df['Average'] = 0

for i in tqdm(range(test_df.shape[0])):
    Xtest = test_df.iloc[i:(i+1)].copy()
    y_pred = lgb_model.predict(Xtest.loc[:, ~Xtest.columns.isin(['Average','WeekEnding_Date'])], 
                               num_iteration=lgb_model.best_iteration_)
    
    test_df.loc[i, 'Average'] = y_pred.ravel()
    df = pd.concat([train_df[train_df['year']==2019], test_df])
    
    window_dict = {
        'Average': [1, 2, 14, 52]
    }
    for col in window_dict.keys():
        for w in window_dict.get(col):
            df[col+'_Lag'+str(w)] = df[col].shift(w)
            df[col+'_LagDiff'+str(w)] = df[col] - df[col+'_Lag'+str(w)]
    
    df.fillna(-1, inplace=True)
    test_df = df[df['WeekEnding_Date']>=pd.Timestamp(datetime.date(2019, 6, 15))].copy()
```
