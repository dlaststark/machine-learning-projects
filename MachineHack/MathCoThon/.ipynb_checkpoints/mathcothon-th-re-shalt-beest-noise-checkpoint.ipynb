## Import libraries

```python
import gc
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras.utils import Sequence
from tensorflow.keras.regularizers import l2
from tensorflow.keras.utils import plot_model
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow_addons.optimizers import AdamW, Lookahead
from tensorflow.keras.layers import LayerNormalization
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Embedding, Concatenate
from tensorflow.keras.layers import Conv1D, PReLU, Flatten
from tensorflow.keras.layers import Dense, Input, Activation
```

## Prepare data for model training

```python
with open("../input/mathcothon-i-shalt-prepareth-data/MathCoThon_Ready_Meatballs.txt", 'rb') as handle: 
    data = handle.read()

processed_data = pickle.loads(data)
train_df = processed_data['train_df']
test_df = processed_data['test_df']
train_df.drop(['Price'], inplace=True, axis=1)

del processed_data
gc.collect()
```

```python
cat_cols = ['Manufacturer','Model','Category','Leather interior','Cylinders',
            'Doors','Wheel','Color','Airbags','Turbo_Engine','Leap_Year',
            'Hybrid_Car','Numbers_in_Model','Drive wheels Front','Drive wheels Rear',
            'Manual_Gear','Automatic_Gear','Continuous_Gear','Fuel Tank Petrol',
            'Fuel Tank Diesel','Fuel Tank Gas','id0','id1','id2','id3','id4','id5','id6','id7']

vec_cols = [col for col in test_df.columns if (col.startswith('tfidf_') or col.startswith('countvec_'))]
num_cols = [col for col in test_df.columns if not(col in cat_cols or col in vec_cols)]

train_df[cat_cols] = train_df[cat_cols].astype(int)
test_df[cat_cols] = test_df[cat_cols].astype(int)
```

```python
scaler = MinMaxScaler().fit(train_df)
temp_x1 = scaler.transform(train_df)
temp_x2 = scaler.transform(test_df)

train_df = pd.DataFrame(temp_x1, columns=test_df.columns, index=train_df.index)
test_df = pd.DataFrame(temp_x2, columns=test_df.columns, index=test_df.index)
```

```python
combined_df = train_df.append(test_df, sort=False, ignore_index=True)
combined_df.head()
```

```python
class SwapNoise(Sequence):

    def __init__(self, num_data, cat_data, vec_data, swap_perc=0.15, batch_size=128, shuffle=True):
        self.batch_size = batch_size
        self.data = num_data
        self.cat_data = cat_data
        self.vec_data = vec_data
        self.swap_perc = swap_perc
        self.shuffle = shuffle
        self.on_epoch_end()


    def __len__(self):
        return self.data.shape[0] // self.batch_size


    def __getitem__(self, index):
        indices = [i for i in range(index*self.batch_size, 
                                    (index+1)*self.batch_size)]
        noisy = self.swap_noise(self.data[indices])
        return [self.cat_data[indices], 
                noisy, 
                self.vec_data[indices]], np.concatenate((self.cat_data[indices],
                                                         self.data[indices],  
                                                         self.vec_data[indices]), axis=1)


    def on_epoch_end(self):
        if self.shuffle == True:
            np.random.shuffle(self.data)


    def get_data(self):
        return self.data


    def swap_noise(self, temp):
        for index in range(temp.shape[0]):            
            num_swaps = np.random.binomial(temp.shape[1], 
                                           self.swap_perc, 
                                           size=1)
            if num_swaps ==0:
                num_swaps=1

            swap_rvs = np.random.normal(size=num_swaps)
            positions_in_row = np.random.choice(a=range(2,temp.shape[1]), 
                                                size=num_swaps, 
                                                replace=False)
            temp[index, positions_in_row] = swap_rvs
        
        return temp
```

```python
mini_batch_size = 128

train_, test_ = train_test_split(combined_df, test_size=0.2)
print("train_: {} \ntest_: {}".format(train_.shape, test_.shape))

Xtrain = SwapNoise(train_[num_cols].values, train_[cat_cols].values, train_[vec_cols].values, 
                   swap_perc=0.15, batch_size=mini_batch_size, shuffle=True)
Xtest = SwapNoise(test_[num_cols].values, test_[cat_cols].values, test_[vec_cols].values, 
                  swap_perc=0.15, batch_size=mini_batch_size, shuffle=True)
```

## Build the model

```python
def autoencoder(cat_features, num_features, vec_features):

    x_input1 = Input(shape=(cat_features, ), name='Categorical Features')
    x_input2 = Input(shape=(num_features, ), name='Numerical Features')
    x_input3 = Input(shape=(vec_features, ), name='Text Vector Features')
    
    xe1 = Embedding(128, 8, embeddings_initializer='he_normal')(x_input1)
    xe1 = LayerNormalization(epsilon=1e-5)(xe1)
    
    xe1 = Conv1D(filters=8, kernel_size=1, 
                 strides=1, padding='same', 
                 kernel_regularizer=l2(0.0005),
                 kernel_initializer='he_normal')(xe1)
    xe1 = LayerNormalization(epsilon=1e-5)(xe1)
    xe1 = PReLU()(xe1)
    xe1 = Flatten()(xe1)

    xe2 = Dense(units=16, kernel_initializer='he_normal', 
                kernel_regularizer=l2(0.0005))(x_input2)
    xe2 = BatchNormalization()(xe2)
    xe2 = PReLU()(xe2)
    
    xe3 = Dense(units=128, kernel_initializer='he_normal', 
                kernel_regularizer=l2(0.0005))(x_input3)
    xe3 = BatchNormalization()(xe3)
    xe3 = PReLU()(xe3)
    
    xe4 = Concatenate()([xe1, xe2, xe3])
    xe4 = BatchNormalization()(xe4)

    embed = Dense(units=256, kernel_initializer='he_normal', 
                  kernel_regularizer=l2(0.0005))(xe4)
    embed = BatchNormalization()(embed)
    embed = PReLU(name='Embedding')(embed)
    
    xd1 = Dense(units=512, kernel_initializer='he_normal', 
                kernel_regularizer=l2(0.0005))(embed)
    xd1 = BatchNormalization()(xd1)
    xd1 = PReLU()(xd1)

    x_output = Dense(units=(cat_features + num_features + vec_features), 
                     kernel_initializer='he_normal')(xd1)

    model = Model(inputs=[x_input1, x_input2, x_input3], outputs=x_output, name='Autoencoder_Model')
    return model
```

```python
tf.random.set_seed(10)

model = autoencoder(len(cat_cols), len(num_cols), len(vec_cols))

model.compile(loss='mean_squared_error',
              optimizer=Lookahead(AdamW(lr=1e-2, 
                                        weight_decay=1e-5, 
                                        clipvalue=700), 
                                  sync_period=10))

early = EarlyStopping(monitor="val_loss", mode="min", 
                      restore_best_weights=True, 
                      patience=15, verbose=1)

reduce_lr = ReduceLROnPlateau(monitor="val_loss", factor=0.25, 
                              min_lr=1e-6, patience=4, 
                              verbose=1, mode='min')

chk_point = ModelCheckpoint('./DAE_model.h5', 
                            monitor='val_loss', verbose=1, 
                            save_best_only=True, mode='min')

model.summary()
```

```python
plot_model(
    model, to_file='./DAE_model.png', 
    show_shapes=True, show_layer_names=True
)
```

```python
history = model.fit(
    Xtrain, 
    batch_size=mini_batch_size,
    epochs=300, 
    verbose=1,
    callbacks=[reduce_lr, early, chk_point], 
    validation_data=Xtest
)
```

```python
plt.figure(figsize=(15,8))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper right')
plt.show()
```
