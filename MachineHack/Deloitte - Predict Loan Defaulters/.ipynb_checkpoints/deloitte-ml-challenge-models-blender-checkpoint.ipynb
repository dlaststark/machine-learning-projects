## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import plotly.express as px
import plotly.figure_factory as ff
```

## Load model predictions

```python
ds = np.load('../input/deloitte-ml-challenge-tree-models/Meta_Features.npz', allow_pickle=True)
y_pred_meta_lgb11, y_pred_final_lgb11 = ds['y_pred_meta_lgb1'], ds['y_pred_final_lgb1']
y_pred_meta_lgb12, y_pred_final_lgb12 = ds['y_pred_meta_lgb2'], ds['y_pred_final_lgb2']
y_pred_meta_cb11, y_pred_final_cb11 = ds['y_pred_meta_cb1'], ds['y_pred_final_cb1']
y_pred_meta_cb12, y_pred_final_cb12 = ds['y_pred_meta_cb2'], ds['y_pred_final_cb2']
y_pred_meta_xgb11, y_pred_final_xgb11 = ds['y_pred_meta_xgb1'], ds['y_pred_final_xgb1']
y_pred_meta_xgb12, y_pred_final_xgb12 = ds['y_pred_meta_xgb2'], ds['y_pred_final_xgb2']
```

```python
ds = np.load('../input/deloitte-ml-challenge-tree-models-v2/Meta_Features.npz', allow_pickle=True)
y_pred_meta_lgb21, y_pred_final_lgb21 = ds['y_pred_meta_lgb1'], ds['y_pred_final_lgb1']
y_pred_meta_lgb22, y_pred_final_lgb22 = ds['y_pred_meta_lgb2'], ds['y_pred_final_lgb2']
y_pred_meta_cb21, y_pred_final_cb21 = ds['y_pred_meta_cb1'], ds['y_pred_final_cb1']
y_pred_meta_cb22, y_pred_final_cb22 = ds['y_pred_meta_cb2'], ds['y_pred_final_cb2']
y_pred_meta_xgb21, y_pred_final_xgb21 = ds['y_pred_meta_xgb1'], ds['y_pred_final_xgb1']
y_pred_meta_xgb22, y_pred_final_xgb22 = ds['y_pred_meta_xgb2'], ds['y_pred_final_xgb2']
```

## Weighted Average Ensemble

```python
sub11, sub12 = y_pred_final_lgb11.ravel(), y_pred_final_lgb12.ravel()
sub13, sub14 = y_pred_final_cb11.ravel(), y_pred_final_cb12.ravel()
sub15, sub16 = y_pred_final_xgb11.ravel(), y_pred_final_xgb12.ravel()

sub21, sub22 = y_pred_final_lgb21.ravel(), y_pred_final_lgb22.ravel()
sub23, sub24 = y_pred_final_cb21.ravel(), y_pred_final_cb22.ravel()
sub25, sub26 = y_pred_final_xgb21.ravel(), y_pred_final_xgb22.ravel()
```

```python
final_sub1 = (((sub14 + sub15 + sub16)/3.0) * 0.05) + (sub12 * 0.1) + (sub13 * 0.6) + (sub11 * 0.25)
final_sub2 = (((sub21 + sub22)/2.0) * 0.25) + (((sub23 + sub24)/2.0) * 0.65) + (((sub25 + sub26)/2.0) * 0.1)
hist_data = [sub11, sub12, sub13, sub14, sub15, sub16, 
             sub21, sub22, sub23, sub24, sub25, sub26,
             final_sub1, final_sub2]
group_labels = ['sub11', 'sub12', 'sub13', 'sub14', 'sub15', 'sub16', 
                'sub21', 'sub22', 'sub23', 'sub24', 'sub25', 'sub26',
                'final_sub1', 'final_sub2']
fig = ff.create_distplot(hist_data, group_labels, bin_size=0.3, show_hist=False, show_rug=False)
fig.show()
```

```python
data = np.corrcoef(hist_data)
fig = px.imshow(data,x=group_labels, y=group_labels)
fig.show()
```

## Create submission files

```python
submission_df = pd.read_csv("../input/deloitte-ml-challenge/submission.csv")
submission_df['Loan Status'] = np.clip(final_sub1, 0, 0.25)
submission_df.to_csv('blend_submission1.csv', index=False)
submission_df.head(10)
```

```python
submission_df = pd.read_csv("../input/deloitte-ml-challenge/submission.csv")
submission_df['Loan Status'] = np.clip(final_sub2, 0, 0.25)
submission_df.to_csv('blend_submission2.csv', index=False)
submission_df.head(10)
```

```python
submission_df = pd.read_csv("../input/deloitte-ml-challenge/submission.csv")
submission_df['Loan Status'] = np.clip((final_sub1 * 0.75) + (final_sub2 * 0.25), 0, 0.25)
submission_df.to_csv('blend_submission3.csv', index=False)
submission_df.head(10)
```
