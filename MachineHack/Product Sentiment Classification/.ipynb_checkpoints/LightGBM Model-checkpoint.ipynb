### Helper functions

```python
def char_is_emoji(character):
    """
        Description: Function to check if input character is an emoji
    """
    return character in emoji.UNICODE_EMOJI
```

```python
def text_has_emoji(text):
    """
        Description: Function to check if input text contains an emoji
    """
    for character in text:
        if character in emoji.UNICODE_EMOJI:
            return True
    return False
```

```python
def deEmojify(inputString):
    """
        Description: Function to remove the emojis in input string
    """
    return inputString.encode('ascii', 'ignore').decode('ascii')
```

```python
def remove_punctuations(text):
    """
        Description: Function to remove any punctuations in input text
    """

    punct =[]
    punct += list(string.punctuation)
    punct += '’'
    punct += '-'
    punct.remove("'")
    
    for punctuation in punct:
        text = text.replace(punctuation, ' ')
    return text
```

```python
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
        
    return input_txt  
```

```python
def nlp(df, field_name):
    """
        Description: Function to perform basic pre-processing in 'Song_Name' field
        
        Parameters:
            df - Input dataframe
            field_name - Field name to be processed
        
        Return:
            df - Processed dataframe
    """
  
    # lowercase everything
    df['token'] = df[field_name].apply(lambda x: x.lower())
    
    # get rid of '\n' from whitespace 
    df['token'] = df['token'].apply(lambda x: x.replace('\n', ' '))
    
    # regex remove hyperlinks
    df['token'] = df['token'].str.replace('http\S+|www.\S+', '', case=False)
    df['token'] = df['token'].apply(lambda x: x.replace('{link}', ''))
    
    # removing '&gt;'
    df['token'] = df['token'].apply(lambda x: x.replace('&gt;', ''))

    # removing '@mention'
    df['token'] = np.vectorize(remove_pattern)(df['token'], "@[\w]*")

    # removing '&quot;'
    df['token'] = df['token'].apply(lambda x: x.replace('&quot;', ''))

    # removing hashtags
    #df['token'] = np.vectorize(remove_pattern)(df['token'], "#[\w]*")
    
    # Checking if emoji in tokens column
    df['emoji'] = df['token'].apply(lambda x: text_has_emoji(x))
    df['emoji'] = df['emoji'].astype(int)
    
    # Removing Emojis from tokens
    df['token'] = df['token'].apply(lambda x: deEmojify(x))
    
    # remove punctuations, numbers and special characters
    df['token'] = df['token'].apply(remove_punctuations)
    df['token'] = df['token'].str.replace("[^a-zA-Z#]", " ")
    
    # remove ' s ' that was created after removing punctuations
    df['token'] = df['token'].apply(lambda x: str(x).replace(" s ", " "))

    # remove short words
    df['token'] = df['token'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>=3 and w.isalpha()]))
    
    return df
```

```python
sp = spacy.load('en_core_web_sm')
lookups = Lookups()
lemm = Lemmatizer(lookups)
```

```python
def lemma_function(text):
    """
        Description: Function to lemmatize the input text
    """

    dummy = []    
    for word in sp(text):
        dummy.append(word.lemma_)
    return ' '.join(dummy)
```

```python
def plot_confusion_matrix(cm, classes):
  
    import itertools

    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion matrix')
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=0)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

#### Feature Engineering

##### Text_ID

```python
combined_df['text_id_odd_even'] = combined_df['Text_ID'].apply(lambda x: 'EVEN' if x%2==0 else 'ODD')
dummy_val = pd.get_dummies(combined_df['text_id_odd_even'], prefix='text_id_odd_even')
combined_df = pd.concat([combined_df, dummy_val], axis=1)
```

```python
combined_df['text_id_0'] = combined_df['Text_ID'].apply(lambda x: int(str(x)[0]))
combined_df['text_id_1'] = combined_df['Text_ID'].apply(lambda x: int(str(x)[1]) if len(str(x))>1 else -1)
combined_df['text_id_2'] = combined_df['Text_ID'].apply(lambda x: int(str(x)[2]) if len(str(x))>2 else -1)
combined_df['text_id_3'] = combined_df['Text_ID'].apply(lambda x: int(str(x)[3]) if len(str(x))>3 else -1)
```

```python
dummy_val = pd.get_dummies(combined_df['text_id_0'], prefix='text_id_0')
combined_df = pd.concat([combined_df, dummy_val], axis=1)

dummy_val = pd.get_dummies(combined_df['text_id_1'], prefix='text_id_1')
combined_df = pd.concat([combined_df, dummy_val], axis=1)

dummy_val = pd.get_dummies(combined_df['text_id_2'], prefix='text_id_2')
combined_df = pd.concat([combined_df, dummy_val], axis=1)

dummy_val = pd.get_dummies(combined_df['text_id_3'], prefix='text_id_3')
combined_df = pd.concat([combined_df, dummy_val], axis=1)
```

```python
combined_df.drop(['Text_ID', 'text_id_odd_even', 'text_id_0', 'text_id_1', 'text_id_2', 'text_id_3'], axis=1, inplace=True)
combined_df.head()
```

##### Product_Type

```python
dummy_val = pd.get_dummies(combined_df['Product_Type'], prefix='Product_Type')
combined_df = pd.concat([combined_df, dummy_val], axis=1)
```

```python
combined_df.drop(['Product_Type'], axis=1, inplace=True)
combined_df.head()
```

##### Product_Description

```python
# Number of words in the "Product_Description" field
combined_df["Prod_Desc_num_words"] = combined_df["Product_Description"].apply(lambda x: len(str(x).split()))

# Number of unique words in the "Product_Description" field
combined_df["Prod_Desc_num_unique_words"] = combined_df["Product_Description"].apply(lambda x: len(set(str(x).split())))

# Number of characters in the "Product_Description" field
combined_df["Prod_Desc_num_chars"] = combined_df["Product_Description"].apply(lambda x: len(str(x)))

# Number of stopwords in the "Product_Description" field
stop = stopwords.words('english')
combined_df["Prod_Desc_num_stopwords"] = combined_df["Product_Description"].apply(lambda x: len([w for w in str(x).lower().split() if w in stop]))

# Number of punctuations in the "Product_Description" field
combined_df["Prod_Desc_num_punctuations"] = combined_df['Product_Description'].apply(lambda x: len([c for c in str(x) if c in string.punctuation]))

# Number of emojis in the "Product_Description" field
combined_df["Prod_Desc_num_emojis"] = combined_df['Product_Description'].apply(lambda x: len([c for c in str(x) if c in emoji.UNICODE_EMOJI]))

# Number of Hashtags in the "Product_Description" field
combined_df['Prod_Desc_num_hastags'] = combined_df['Product_Description'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))

# Number of @ in the "Product_Description" field
combined_df['Prod_Desc_num_@'] = combined_df['Product_Description'].apply(lambda x: len([x for x in x.split() if x.startswith('@mention')]))

# Number of {link} in the "Product_Description" field
combined_df['Prod_Desc_num_link'] = combined_df['Product_Description'].apply(lambda x: len([x for x in x.split() if x.startswith('{link}')]))

# Number of '&quot;' in the "Product_Description" field
combined_df['Prod_Desc_num_quot'] = combined_df['Product_Description'].apply(lambda x: len([x for x in x.split() if x.startswith('&quot;')]))

# Number of numerics in the "Product_Description" field
combined_df['Prod_Desc_num_numerics'] = combined_df['Product_Description'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))

# Number of title case words in the "Product_Description" field
combined_df["Prod_Desc_num_words_upper"] = combined_df["Product_Description"].apply(lambda x: len([w for w in str(x).split() if w.isupper()]))

# Number of title case words in the "Product_Description" field
combined_df["Prod_Desc_num_words_title"] = combined_df["Product_Description"].apply(lambda x: len([w for w in str(x).split() if w.istitle()]))

# Average length of the words in the "Product_Description" field
combined_df["Prod_Desc_mean_word_len"] = combined_df["Product_Description"].apply(lambda x: np.mean([len(w) for w in str(x).split()]))
```

```python
combined_df["X1"] = combined_df["Prod_Desc_num_unique_words"] / combined_df["Prod_Desc_num_words"]
combined_df["X2"] = combined_df["Prod_Desc_num_stopwords"] / combined_df["Prod_Desc_num_words"]
combined_df["X3"] = combined_df["Prod_Desc_num_punctuations"] / combined_df["Prod_Desc_num_words"]
combined_df["X4"] = combined_df["Prod_Desc_num_emojis"] / combined_df["Prod_Desc_num_words"]
combined_df["X5"] = combined_df["Prod_Desc_num_hastags"] / combined_df["Prod_Desc_num_words"]
combined_df["X6"] = combined_df["Prod_Desc_num_@"] / combined_df["Prod_Desc_num_words"]
combined_df["X7"] = combined_df["Prod_Desc_num_numerics"] / combined_df["Prod_Desc_num_words"]
combined_df["X8"] = combined_df["Prod_Desc_num_words_upper"] / combined_df["Prod_Desc_num_words"]
combined_df["X9"] = combined_df["Prod_Desc_num_words_title"] / combined_df["Prod_Desc_num_words"]
combined_df["X10"] = combined_df["Prod_Desc_num_link"] / combined_df["Prod_Desc_num_words"]
combined_df["X11"] = combined_df["Prod_Desc_num_quot"] / combined_df["Prod_Desc_num_words"]
```

```python
# Apply basis pre-processing on 'Product_Description' field
combined_df = nlp(combined_df, "Product_Description")

# Lemmatize the 'token' field
combined_df['lemmatized'] = combined_df['token'].apply(lambda x: lemma_function(x))

# Remove any stop-words in 'lemmatized' field
stop = stopwords.words('english')
combined_df['Processed_Desc'] = combined_df['lemmatized'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))

combined_df.head()
```

```python
# Create two new columns 'Subjectivity' & 'Polarity'
combined_df['Subjectivity'] = combined_df['token'].apply(lambda text: TextBlob(text).sentiment.subjectivity)
combined_df['Polarity'] = combined_df['token'].apply(lambda text: TextBlob(text).sentiment.polarity)
```

```python
# Create new column to calculate "Analysis" based on "Polarity"
combined_df['Analysis'] = combined_df['Polarity'].apply(lambda score: 'Negative' if score<0 else 'Neutral' if score==0 else 'Positive')
dummy_val = pd.get_dummies(combined_df['Analysis'], prefix='Analysis')
combined_df = pd.concat([combined_df, dummy_val], axis=1)
```

```python
# Drop redundant fields
combined_df.drop(['Product_Description', 'token', 'lemmatized', 'Analysis'], inplace=True, axis=1)
combined_df.head()
```

### Hyperparameter search using Optuna

```python
# Convert Ytest to one-hot encoding
df = pd.DataFrame(Ytest, columns=["class"])
Ytest_tmp = pd.get_dummies(df['class']).values
```

```python
def objective(trial):
    
    params = {
        "objective": "multiclass",
        "metric": "multi_logloss",
        "num_class": 4,
        "verbosity": -1,
        "boosting_type": "gbdt",
        "is_unbalance": True,
        "device_type": "gpu",
        "learning_rate": trial.suggest_loguniform("learning_rate", 1e-2, 1e-1),
        "lambda_l2": trial.suggest_loguniform("lambda_l2", 1e-4, 1.0),
        "num_leaves": trial.suggest_int("num_leaves", 50, 1000),
        "max_depth": trial.suggest_int("max_depth", 8, 16),
        "feature_fraction": trial.suggest_uniform("feature_fraction", 0.5, 1.0),
        "bagging_fraction": trial.suggest_uniform("bagging_fraction", 0.5, 1.0),
        "bagging_freq": trial.suggest_int("bagging_freq", 5, 20),
        "min_child_samples": trial.suggest_int("min_child_samples", 1, 20)
    }

    lgtrain = lgb.Dataset(Xtrain, label=Ytrain.ravel())
    lgvalidation = lgb.Dataset(Xvalidation, label=Yvalidation.ravel())

    model = lgb.train(params, lgtrain, valid_sets=[lgvalidation], 
                      num_boost_round=5000, early_stopping_rounds=200, 
                      verbose_eval=False)
    
    y_pred = model.predict(Xtest, num_iteration=model.best_iteration)
    loss = log_loss(Ytest_tmp, y_pred)
    return loss
```

```python
study = optuna.create_study()
study.optimize(objective, n_trials=500)
```

```python
print("Number of finished trials: {}".format(len(study.trials)))

print("Best trial:")
trial = study.best_trial

print("Value: {}".format(trial.value))

print("Params: ")
for key, value in trial.params.items():
    print(" {}: {}".format(key, value))
```

### Build the model

```python
# Define model hyperparameters
params = {}
params["objective"] = 'multiclass'
params["metric"] = 'multi_logloss'
params["num_class"] = 4
params["is_unbalance"] = True
params["boosting"] = 'gbdt'
params["device_type"] = 'gpu'
params["learning_rate"] = 0.0317
params["lambda_l2"] = 0.0142
params["num_leaves"] = 183
params["max_depth"] = 9
params["feature_fraction"] = 0.784
params["bagging_fraction"] = 0.925
params["bagging_freq"] = 11
params["min_data_in_leaf"] = 14
params["verbosity"] = -1
num_rounds = 5000
```

```python
Xtrain_tmp = np.concatenate((Xtrain, Xvalidation), axis=0)
Ytrain_tmp = np.concatenate((Ytrain, Yvalidation), axis=0)
```

```python
# Define K-fold cross validation test harness
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=20)
y_pred = 0
```

```python
# Train the model using K-fold
counter = 0

for train, val in kfold.split(Xtrain_tmp, Ytrain_tmp):
    counter += 1

    train_x, train_y = Xtrain_tmp[train], Ytrain_tmp[train]
    val_x, val_y = Xtrain_tmp[val], Ytrain_tmp[val]

    lgtrain = lgb.Dataset(train_x, label=train_y.ravel())
    lgvalidation = lgb.Dataset(val_x, label=val_y.ravel())

    model = lgb.train(params, lgtrain, num_rounds, 
                      valid_sets=[lgtrain, lgvalidation], early_stopping_rounds=200, 
                      verbose_eval=100)
    
    pred = model.predict(Xtest, num_iteration=model.best_iteration)
    y_pred = y_pred + pred

y_pred = y_pred/float(counter)
```

### Validate the model

```python
loss = log_loss(Ytest_tmp, y_pred)
y_pred_binary = np.array([np.argmax(y_pred, axis=1)]).T
f1 = f1_score(Ytest, y_pred_binary, average='weighted')
print('Overall log_loss:', loss)
print('Overall F1-Score:', f1)
```

```python
print(classification_report(Ytest, y_pred_binary))
```

```python
cnf_matrix = confusion_matrix(Ytest, y_pred_binary, labels=[0, 1, 2, 3])
np.set_printoptions(precision=2)
plt.figure(figsize=(12,7))
plot_confusion_matrix(cnf_matrix, classes=[0, 1, 2])
```

### Train model on entire data

```python
Xtrain_tmp = np.concatenate((Xtrain, Xvalidation, Xtest), axis=0)
Ytrain_tmp = np.concatenate((Ytrain, Yvalidation, Ytest), axis=0)
```

```python
# Define 5-fold cross validation test harness
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=30)
y_pred = 0
```

```python
# Train the model using K-fold
counter = 0

for train, val in kfold.split(Xtrain_tmp, Ytrain_tmp):
    counter += 1

    train_x, train_y = Xtrain_tmp[train], Ytrain_tmp[train]
    val_x, val_y = Xtrain_tmp[val], Ytrain_tmp[val]

    lgtrain = lgb.Dataset(train_x, label=train_y.ravel())
    lgvalidation = lgb.Dataset(val_x, label=val_y.ravel())

    model = lgb.train(params, lgtrain, num_rounds, 
                      valid_sets=[lgtrain, lgvalidation], early_stopping_rounds=200, 
                      verbose_eval=100)
    
    pred = model.predict(Xpredict, num_iteration=model.best_iteration)
    y_pred = y_pred + pred

y_pred = y_pred/float(counter)
```
