```python
train_df['rel_year'] = train_df['Timestamp'].map(lambda x: pd.to_datetime(x).year)
train_df['rel_quarter'] = train_df['Timestamp'].map(lambda x: pd.to_datetime(x).quarter)
train_df['rel_month'] = train_df['Timestamp'].map(lambda x: pd.to_datetime(x).month)
train_df['rel_week'] = train_df['Timestamp'].map(lambda x: pd.to_datetime(x).week)
train_df['rel_day_year'] = train_df['Timestamp'].map(lambda x: pd.to_datetime(x).dayofyear)
train_df['rel_day_month'] = train_df['Timestamp'].map(lambda x: pd.to_datetime(x).day)
train_df['rel_day_week'] = train_df['Timestamp'].map(lambda x: pd.to_datetime(x).dayofweek)
```

```python
train_df.head(20)
```

```python
train_df['Likes'] = train_df['Likes'].map(lambda x: x.replace(",", ""))
```

```python
train_df['Likes_new'] = (train_df.Likes.replace(r'[KM]+$', '', regex=True).astype(float) * 
                         train_df.Likes.str.extract(r'[\d\.]+([KM]+)', expand=False)
                         .fillna(1)
                         .replace(['K','M'], [10**3, 10**6]).astype(int))
```

```python
train_df['Popularity'] = train_df['Popularity'].map(lambda x: x.replace(",", ""))
```
