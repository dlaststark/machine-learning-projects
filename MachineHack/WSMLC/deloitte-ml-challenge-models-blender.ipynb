```python
import numpy as np
import pandas as pd
import matplotlib as plt
import plotly.express as px
import plotly.figure_factory as ff
```

```python
ds = np.load('../input/deloitte-ml-challenge-tree-models/Meta_Features.npz', allow_pickle=True)
y_pred_meta_lgb1, sub1 = ds['y_pred_meta_lgb1'], ds['y_pred_final_lgb1'].ravel()
y_pred_meta_lgb2, sub2 = ds['y_pred_meta_lgb2'], ds['y_pred_final_lgb2'].ravel()
y_pred_meta_cb1, sub3 = ds['y_pred_meta_cb1'], ds['y_pred_final_cb1'].ravel()
y_pred_meta_cb2, sub4 = ds['y_pred_meta_cb2'], ds['y_pred_final_cb2'].ravel()
y_pred_meta_xgb1, sub5 = ds['y_pred_meta_xgb1'], ds['y_pred_final_xgb1'].ravel()
y_pred_meta_xgb2, sub6 = ds['y_pred_meta_xgb2'], ds['y_pred_final_xgb2'].ravel()
```

```python
final_sub = (((sub4 + sub5 + sub6 + sub2)/4.0) * 0.1) + (sub3 * 0.65) + (sub1 * 0.25)
hist_data = [sub1, sub2, sub3, sub4, sub5, sub6, final_sub]
group_labels = ['sub1', 'sub2', 'sub3', 'sub4', 'sub5', 'sub6', 'final_sub']
fig = ff.create_distplot(hist_data, group_labels, bin_size=0.3, show_hist=False, show_rug=False)
fig.show()
```

```python
data = np.corrcoef([sub1, sub2, sub3, sub4, sub5, sub6, final_sub])
fig = px.imshow(data,x=group_labels, y=group_labels)
fig.show()
```

```python
submission_df = pd.read_csv("../input/deloitte-ml-challenge/submission.csv")
submission_df['Loan Status'] = np.clip(final_sub, 0, 0.25)
submission_df.to_csv('blend_submission.csv', index=False)
submission_df.head(10)
```
