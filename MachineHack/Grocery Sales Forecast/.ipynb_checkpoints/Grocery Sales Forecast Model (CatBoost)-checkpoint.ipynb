## Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.stattools import adfuller
from catboost import CatBoostRegressor
import optuna
```

## Read training data

```python
train_df = pd.read_csv("Dataset/Train.csv", index_col='Day')
train_df.head()
```

## Exploratory Data Analysis

```python
def test_stationarity(timeseries):
    '''
    Check stationarity of time series data
    '''
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
plt.figure(figsize=(15,8))
plt.grid()
test_stationarity(train_df)
```

## Data Preprocessing

#### Create train and test dataset

```python
window_size = 60
test_size = 60
```

```python
train_x, test_x = train_df[:-test_size], train_df[-(window_size + test_size):]
print("train_x: {} \ntest_x: {}".format(train_x.shape, test_x.shape))
```

#### Normalize dataset

```python
scaler = MinMaxScaler().fit(train_x)
train_x = scaler.transform(train_x)
test_x = scaler.transform(test_x)
```

#### Plot train/test dataset division

```python
plt.figure(figsize=(15,8))
plt.plot(np.arange(0, len(train_x)), train_x, 'g', label="train data")
plt.plot(np.arange(len(train_x), len(train_x) + len(test_x)), test_x, 'r', label="test data")
plt.legend()
plt.grid()
plt.show()
```

#### Time-series data preparation using rolling window

```python
def create_dataset(dataset, time_steps=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-time_steps):
        a = dataset[i:(i+time_steps), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_steps, 0])
    return np.array(dataX), np.array(dataY)
```

```python
Xtrain = train_x.copy()
Xtest = test_x.copy()

# Reshape data to [samples, time_steps, n_features]
Xtrain, Ytrain = create_dataset(Xtrain, window_size)
Xtest, Ytest = create_dataset(Xtest, window_size)

Ytrain = np.array([Ytrain]).T
Ytest = np.array([Ytest]).T

print("Xtrain: {} \nYtrain: {}".format(Xtrain.shape, Ytrain.shape))
print("Xtest: {} \nYtest: {}".format(Xtest.shape, Ytest.shape))
```

## Hyperparameter search using Optuna

```python
def objective(trial):
        
    model = CatBoostRegressor(
        objective='RMSE',
        num_boost_round=5000,
        random_seed=20,
        learning_rate=trial.suggest_loguniform("learning_rate", 1e-2, 1e-1),
        l2_leaf_reg=trial.suggest_loguniform("l2_leaf_reg", 1e-4, 1.0),
        #max_leaves=trial.suggest_int("max_leaves", 31, 150),
        max_depth=trial.suggest_int("max_depth", 6, 15),
        bagging_temperature=trial.suggest_uniform("bagging_temperature", 0.5, 1.0),
        rsm=trial.suggest_uniform("rsm", 0.5, 1.0),
        min_data_in_leaf=trial.suggest_int("min_data_in_leaf", 1, 20)
    )

    model.fit(Xtrain, Ytrain, eval_set=(Xtest, Ytest), early_stopping_rounds=200, verbose_eval=False)
    y_pred = model.predict(Xtest)
    rmse = np.sqrt(mean_squared_error(Ytest, y_pred))
    score = 100 * max(0, 1 - rmse)

    return score
```

```python
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=50)
```

```python
print("Number of finished trials: {}".format(len(study.trials)))

print("Best trial:")
trial = study.best_trial

print("  Value: {}".format(trial.value))

print("Params: ")
for key, value in trial.params.items():
    print("  {}: {}".format(key, value))
```
