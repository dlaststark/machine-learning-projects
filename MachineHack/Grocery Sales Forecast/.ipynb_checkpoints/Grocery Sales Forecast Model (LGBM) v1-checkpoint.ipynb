## Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.stattools import adfuller
import lightgbm as lgb
import optuna
```

## Read training data

```python
train_df = pd.read_csv("Dataset/Train.csv", index_col='Day')
train_df.head()
```

## Exploratory Data Analysis

```python
def test_stationarity(timeseries):
    '''
    Check stationarity of time series data
    '''
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
plt.figure(figsize=(15,8))
plt.grid()
test_stationarity(train_df)
```

## Data Preprocessing

#### Create train and test dataset

```python
window_size = 60
validation_size = 30
test_size = 30
```

```python
train_x = train_df[:-(validation_size + test_size)]
validation_x = train_df[(len(train_x) - window_size):(len(train_x) + validation_size)]
test_x = train_df[-(window_size + test_size):]
print("train_x: {} \nvalidation_x: {} \ntest_x: {}".format(train_x.shape, 
                                                           validation_x.shape, 
                                                           test_x.shape))
```

#### Normalize dataset

```python
scaler = MinMaxScaler().fit(train_x)
train_x = scaler.transform(train_x)
validation_x = scaler.transform(validation_x)
test_x = scaler.transform(test_x)
```

#### Plot train/test dataset division

```python
plt.figure(figsize=(15,8))
plt.plot(np.arange(0, len(train_x)), train_x, 'g', label="train")
plt.plot(np.arange(len(train_x), len(train_x) + len(validation_x)), validation_x, 'r', label="validation")
plt.plot(np.arange(len(train_x) + len(validation_x), len(train_x) + len(validation_x) + len(test_x)), test_x, 'b', label="test")
plt.legend()
plt.grid()
plt.show()
```

#### Time-series data preparation using rolling window

```python
def create_dataset(dataset, time_steps=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-time_steps):
        a = dataset[i:(i+time_steps), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_steps, 0])
    return np.array(dataX), np.array(dataY)
```

```python
Xtrain = train_x.copy()
Xvalidation = validation_x.copy()
Xtest = test_x.copy()

# Reshape data to [samples, time_steps, n_features]
Xtrain, Ytrain = create_dataset(Xtrain, window_size)
Xvalidation, Yvalidation = create_dataset(Xvalidation, window_size)
Xtest, Ytest = create_dataset(Xtest, window_size)

Ytrain = np.array([Ytrain]).T
Yvalidation = np.array([Yvalidation]).T
Ytest = np.array([Ytest]).T

print("Xtrain: {} \nYtrain: {}".format(Xtrain.shape, Ytrain.shape))
print("Xvalidation: {} \nYvalidation: {}".format(Xvalidation.shape, Yvalidation.shape))
print("Xtest: {} \nYtest: {}".format(Xtest.shape, Ytest.shape))
```

## Hyperparameter search using Optuna

```python
def objective(trial):
    
    dtrain = lgb.Dataset(Xtrain, label=Ytrain.ravel())
    dtest = lgb.Dataset(Xvalidation, label=Yvalidation.ravel())

    param = {
        "objective": "regression",
        "metric": "rmse",
        "verbosity": -1,
        "boosting_type": "gbdt",
        "learning_rate": trial.suggest_loguniform("learning_rate", 1e-2, 1e-1),
        "lambda_l2": trial.suggest_loguniform("lambda_l2", 1e-4, 1.0),
        "num_leaves": trial.suggest_int("num_leaves", 31, 150),
        "max_depth": trial.suggest_int("max_depth", 7, 20),
        "feature_fraction": trial.suggest_uniform("feature_fraction", 0.5, 1.0),
        "bagging_fraction": trial.suggest_uniform("bagging_fraction", 0.5, 1.0),
        "bagging_freq": trial.suggest_int("bagging_freq", 5, 20),
        "min_child_samples": trial.suggest_int("min_child_samples", 1, 20),
    }

    gbm = lgb.train(param, dtrain, valid_sets=[dtest], num_boost_round=5000, early_stopping_rounds=200, verbose_eval=False)
    y_pred = gbm.predict(Xtest)
    rmse = np.sqrt(mean_squared_error(Ytest, y_pred))
    score = 100 * max(0, 1 - rmse)

    return score
```

```python
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=100)
```

```python
print("Number of finished trials: {}".format(len(study.trials)))

print("Best trial:")
trial = study.best_trial

print("  Value: {}".format(trial.value))

print("Params: ")
for key, value in trial.params.items():
    print("  {}: {}".format(key, value))
```

## Build the time-series model

#### Define model hyperparameters

```python
params = {}
params["objective"] = 'regression'
params['metric'] = 'rmse'
params['boosting'] = 'gbdt'
params["max_depth"] = 15
params["num_leaves"] = 131
params["learning_rate"] = 0.024
params["bagging_fraction"] = 0.694
params["feature_fraction"] = 0.593
params["bagging_freq"] = 15
params["bagging_seed"] = 10
params["lambda_l2"] = 0.001
params["min_data_in_leaf"] = 18
params["verbosity"] = -1
num_rounds = 5000
```

#### Prepare dataset for LightGBM

```python
lgtrain = lgb.Dataset(Xtrain, label=Ytrain.ravel())
lgtest = lgb.Dataset(Xvalidation, label=Yvalidation.ravel())
```

#### Train the LGBM model

```python
model = lgb.train(params, lgtrain, num_rounds, valid_sets=[lgtest], early_stopping_rounds=200, verbose_eval=1000)
```

#### Make predictions on test dataset

```python
y_pred = model.predict(Xtest)
```

#### Plot the predictions on test dataset

```python
plt.figure(figsize=(15,8))
plt.plot(np.arange(0, len(Ytrain)), Ytrain, 'g', label="history")
plt.plot(np.arange(len(Ytrain), len(Ytrain) + len(Ytest)), Ytest, marker='.', label="true data")
plt.plot(np.arange(len(Ytrain), len(Ytrain) + len(Ytest)), y_pred, 'r', label="prediction")
plt.ylabel('Value')
plt.xlabel('Time Step')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.plot(Ytest, marker='.', label="true data")
plt.plot(y_pred, 'r', label="prediction")
plt.ylabel('Value')
plt.xlabel('Time Step')
plt.legend()
plt.grid()
plt.show()
```

#### Get RMSE value on test dataset predictions

```python
np.sqrt(mean_squared_error(Ytest, y_pred))
```

## Train the model on entire training data

#### Dataset preparation

```python
train_x = train_df.values
scaler = MinMaxScaler().fit(train_x)
train_x = scaler.transform(train_x)

# Reshape data to [samples, time_steps, n_features]
Xtrain, Ytrain = create_dataset(train_x, window_size)
Ytrain = np.array([Ytrain]).T
print("Xtrain: {} \nYtrain: {}".format(Xtrain.shape, Ytrain.shape))
```

#### Train model on entire dataset

```python
lgtrain = lgb.Dataset(Xtrain, label=Ytrain.ravel())
model = lgb.train(params, lgtrain, num_rounds, verbose_eval=1000)
```

#### Make predictions for future 90 days

```python
predictions = train_x[-window_size:].ravel()

for i in range(90):
    Xpred = np.array([predictions[-window_size:]])
    y_pred = model.predict(Xpred)
    predictions = np.append(predictions, y_pred)

Ypred = predictions[-90:]
```

#### Plot future predictions

```python
plt.figure(figsize=(15,8))
plt.plot(np.arange(0, len(Ytrain)), Ytrain, 'g', label="history")
plt.plot(np.arange(len(Ytrain), len(Ytrain) + len(Ypred)), Ypred, 'r', label="prediction")
plt.ylabel('Value')
plt.xlabel('Time Step')
plt.legend()
plt.grid()
plt.show()
```

## Create submission file

```python
final_pred = np.array([Ypred]).T
final_pred = scaler.inverse_transform(final_pred)
print(final_pred.shape)
```

```python
submit_df = pd.DataFrame(final_pred, columns=['GrocerySales'])
submit_df['GrocerySales'] = submit_df['GrocerySales'].astype(float).round(2)
submit_df.head()
```
