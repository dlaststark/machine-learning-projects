```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from tensorflow.keras.layers import Input, Dense, Dropout, LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import RootMeanSquaredError
```

```python
from learningratefinder import LearningRateFinder
from clr_callback import CyclicLR
```

```python
train_df = pd.read_csv("Dataset/Train.csv", index_col='Day')
print(train_df.shape)
train_df.head(10)
```

```python
ts = train_df['GrocerySales']
plt.figure(figsize=(15,8))
plt.grid()
plt.plot(ts)
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
ts_log = ts.copy()
plt.figure(figsize=(15,8))
plt.grid()
plt.plot(ts_log)
```

```python
moving_avg = ts_log.rolling(10).mean()
plt.figure(figsize=(15,8))
plt.grid()
plt.plot(ts_log)
plt.plot(moving_avg, color='red')
```

```python
ts_log_moving_avg_diff = ts_log - moving_avg
ts_log_moving_avg_diff.head(15)
```

```python
ts_log_moving_avg_diff.dropna(inplace=True)
plt.figure(figsize=(15,8))
plt.grid()
test_stationarity(ts_log_moving_avg_diff)
```

```python
window_size = 10
test_size = 48
```

```python
train_x, test_x = np.array([ts_log_moving_avg_diff[:-test_size]]).T, np.array([ts_log_moving_avg_diff[-(window_size + test_size):]]).T
print(train_x.shape)
print(test_x.shape)
```

```python
scaler = MinMaxScaler().fit(train_x)
train_x = scaler.transform(train_x)
test_x = scaler.transform(test_x)
```

```python
plt.figure(figsize=(15,8))
plt.plot(np.arange(0, len(train_x)), train_x, 'g', label="train data")
plt.plot(np.arange(len(train_x), len(train_x) + len(test_x)), test_x, 'r', label="test data")
plt.legend()
plt.grid()
plt.show()
```

```python
def create_dataset(dataset, time_steps=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-time_steps):
        a = dataset[i:(i+time_steps), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_steps, 0])
    return np.array(dataX), np.array(dataY)
```

```python
Xtrain = train_x.copy()
Xtest = test_x.copy()

# reshape to [samples, time_steps, n_features]
Xtrain, Ytrain = create_dataset(Xtrain, window_size)
Xtest, Ytest = create_dataset(Xtest, window_size)

print(Xtrain.shape, Ytrain.shape)
print(Xtest.shape, Ytest.shape)
```

```python
# reshape input to be [samples, time steps, features]
Xtrain = np.reshape(Xtrain, (Xtrain.shape[0], Xtrain.shape[1], 1))
Xtest = np.reshape(Xtest, (Xtest.shape[0], Xtest.shape[1], 1))

print(Xtrain.shape, Ytrain.shape)
print(Xtest.shape, Ytest.shape)
```

```python
simple_lstm_model = Sequential([
    LSTM(128, activation='relu', return_sequences=True, dropout=0.25, input_shape=(Xtrain.shape[1], Xtrain.shape[2])),
    LSTM(128, activation='relu', dropout=0.25),
    Dense(64, activation='relu'),
    Dropout(0.25),
    Dense(64, activation='relu'),
    Dropout(0.25),
    Dense(1)
])

simple_lstm_model.compile(optimizer=Adam(lr=1e-3), loss='mse', metrics=[RootMeanSquaredError(name='rmse')])
simple_lstm_model.summary()
```

```python
# Learning rate Scheduler
min_lr = 1e-5
max_lr = 1e-3
mini_batch_size = 32

step_size = 4 * (Xtrain.shape[0] // mini_batch_size)
clr_method = 'triangular2'

clr = CyclicLR(base_lr=min_lr, max_lr=max_lr, mode=clr_method, step_size=step_size)
```

```python
# Learning Rate Finder
lrf = LearningRateFinder(simple_lstm_model)
lrf.find((Xtrain, Ytrain),
         startLR=1e-10, endLR=1e-1,
         stepsPerEpoch=np.ceil((len(Xtrain) / float(mini_batch_size))),
         batchSize=mini_batch_size)
lrf.plot_loss()
plt.grid()
plt.show()
```

```python
history = simple_lstm_model.fit(
    Xtrain, Ytrain, 
    epochs=50, 
    batch_size=mini_batch_size, 
    validation_split=0.1, 
    callbacks=[clr], 
    workers=5, 
    verbose=1, 
    shuffle=False
)
```

```python
plt.figure(figsize=(15,8))
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.grid()
plt.legend()
```

```python
plt.figure(figsize=(15,8))
plt.plot(history.history['rmse'], label='rmse')
plt.plot(history.history['val_rmse'], label='val_rmse')
plt.grid()
plt.legend()
```

```python
y_pred = simple_lstm_model.predict(Xtest)
```

```python
plt.figure(figsize=(15,8))
plt.plot(np.arange(0, len(Ytrain)), Ytrain, 'g', label="history")
plt.plot(np.arange(len(Ytrain), len(Ytrain) + len(Ytest)), Ytest, marker='.', label="true")
plt.plot(np.arange(len(Ytrain), len(Ytrain) + len(Ytest)), y_pred, 'r', label="prediction")
plt.ylabel('Value')
plt.xlabel('Time Step')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.plot(Ytest, marker='.', label="true")
plt.plot(y_pred, 'r', label="prediction")
plt.ylabel('Value')
plt.xlabel('Time Step')
plt.legend()
plt.grid()
plt.show()
```
