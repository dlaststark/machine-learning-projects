## Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.stattools import adfuller
import lightgbm as lgb
import optuna
```

## Read training data

```python
train_df = pd.read_csv("Dataset/Train.csv", index_col='Day')
train_df.head()
```

## Exploratory Data Analysis

```python
def test_stationarity(timeseries):
    '''
    Check stationarity of time series data
    '''
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
plt.figure(figsize=(15,8))
plt.grid()
test_stationarity(train_df)
```

## Data Preprocessing

#### Normalize training data

```python
train = train_df.values
train.shape
```

```python
scaler = MinMaxScaler().fit(train)
train = scaler.transform(train)
```

#### Prepare dataset for time-series model

```python
window_size = 60
```

```python
def prepare_data(sales_list, item=None):
    
    if item is not None:
        sales_list = sales_list[1:]
        sales_list = np.append(sales_list, item)
        
    df = pd.DataFrame(np.array([sales_list]).T, columns=["Sales"])
    df["cum_mean"] = df['Sales'].expanding().mean()
    df["cum_std"] = df['Sales'].expanding().std()
    df["cum_std"] = df["cum_std"].fillna(0)
    
    for i in range(1, window_size+1):
        if i == window_size:
            df['target'] = df['Sales'].shift(-i)
        else:
            df['sale-lag-'+str(i)] = df['Sales'].shift(-i)
            df['sale-lag-diff-'+str(i)] = df['Sales'] - df['Sales'].shift(-i)
    
        if i < window_size:
            df['mean-lag-'+str(i)] = df['cum_mean'].shift(-i)
            df['std-lag-'+str(i)] = df['cum_std'].shift(-i)
    
    df.dropna(inplace=True)
    df.drop(['cum_mean', 'cum_std'], inplace=True, axis=1)
    
    return sales_list, df
```

```python
def prepare_data(sales_list, item=None):
    
    if item is not None:
        sales_list = sales_list[1:]
        sales_list = np.append(sales_list, item)
        
    df = pd.DataFrame(np.array([sales_list]).T, columns=["Sales"])
    
    for i in range(1, window_size+1):
        if i == window_size:
            df['target'] = df['Sales'].shift(-i)
        else:
            df['sale-lag-'+str(i)] = df['Sales'].shift(-i)
            df['sale-lag-diff-'+str(i)] = df['Sales'] - df['Sales'].shift(-i)
    
    df.dropna(inplace=True)   
    return sales_list, df
```

```python
sales_list = train.ravel()
_, sales_df = prepare_data(sales_list)
```

```python
sales_df.head()
```

```python
train_y = np.array([sales_df['target'].values]).T
sales_df.drop(['target'], inplace=True, axis=1)
train_x = sales_df.values
print("train_x: {} \ntrain_y: {}".format(train_x.shape, train_y.shape))
```

#### Create train and test datasets

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(train_x, train_y, test_size=0.05, shuffle=False, random_state=10)
print("Xtrain: {} \nYtrain: {}".format(Xtrain.shape, Ytrain.shape))
print("Xtest: {} \nYtest: {}".format(Xtest.shape, Ytest.shape))
```

#### Plot train/test dataset division

```python
plt.figure(figsize=(15,8))
plt.plot(np.arange(0, len(Ytrain)), Ytrain, 'g', label="train")
plt.plot(np.arange(len(Ytrain), len(Ytrain) + len(Ytest)), Ytest, 'r', label="test")
plt.legend()
plt.grid()
plt.show()
```

## Hyperparameter search using Optuna

```python
def objective(trial):
    
    dtrain = lgb.Dataset(Xtrain, label=Ytrain.ravel())
    dtest = lgb.Dataset(Xtest, label=Ytest.ravel())

    param = {
        "objective": "regression",
        "metric": "rmse",
        "verbosity": -1,
        "boosting_type": "gbdt",
        "learning_rate": trial.suggest_loguniform("learning_rate", 1e-2, 1e-1),
        "lambda_l2": trial.suggest_loguniform("lambda_l2", 1e-4, 1.0),
        "num_leaves": trial.suggest_int("num_leaves", 31, 200),
        "max_depth": trial.suggest_int("max_depth", 7, 25),
        "feature_fraction": trial.suggest_uniform("feature_fraction", 0.5, 1.0),
        "bagging_fraction": trial.suggest_uniform("bagging_fraction", 0.5, 1.0),
        "bagging_freq": trial.suggest_int("bagging_freq", 5, 20),
        "min_child_samples": trial.suggest_int("min_child_samples", 1, 15),
    }

    gbm = lgb.train(param, dtrain, valid_sets=[dtest], num_boost_round=5000, early_stopping_rounds=200, verbose_eval=False)
    y_pred = gbm.predict(Xtest, num_iteration=gbm.best_iteration)
    rmse = np.sqrt(mean_squared_error(Ytest, y_pred))
    score = 100 * max(0, 1 - rmse)

    return score
```

```python
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=100)
```

```python
print("Number of finished trials: {}".format(len(study.trials)))

print("Best trial:")
trial = study.best_trial

print("  Value: {}".format(trial.value))

print("Params: ")
for key, value in trial.params.items():
    print("  {}: {}".format(key, value))
```

## Build the time-series model

#### Define model hyperparameters

```python
params = {}
params["objective"] = 'regression'
params['metric'] = 'rmse'
params['boosting'] = 'gbdt'
params["max_depth"] = 20
params["num_leaves"] = 89
params["learning_rate"] = 0.056
params["bagging_fraction"] = 0.54
params["feature_fraction"] = 0.85
params["bagging_freq"] = 11
params["bagging_seed"] = 10
params["lambda_l2"] = 0.0006
params["min_data_in_leaf"] = 3
params["verbosity"] = -1
num_rounds = 5000
```

#### Prepare dataset for LightGBM

```python
lgtrain = lgb.Dataset(Xtrain, label=Ytrain.ravel())
lgtest = lgb.Dataset(Xtest, label=Ytest.ravel())
```

#### Train the LGBM model

```python
model = lgb.train(params, lgtrain, num_rounds, valid_sets=[lgtest], early_stopping_rounds=200, verbose_eval=1000)
```

#### Make predictions on test dataset

```python
y_pred = model.predict(Xtest, num_iteration=model.best_iteration)
```

#### Plot the predictions on test dataset

```python
plt.figure(figsize=(15,8))
plt.plot(np.arange(0, len(Ytrain)), Ytrain, 'g', label="history")
plt.plot(np.arange(len(Ytrain), len(Ytrain) + len(Ytest)), Ytest, marker='.', label="true data")
plt.plot(np.arange(len(Ytrain), len(Ytrain) + len(y_pred)), y_pred, 'r', label="prediction")
plt.ylabel('Value')
plt.xlabel('Time Step')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.plot(Ytest, marker='*', label="true data")
plt.plot(y_pred, marker='o', label="prediction")
plt.ylabel('Value')
plt.xlabel('Time Step')
plt.legend()
plt.grid()
plt.show()
```

#### Get RMSE value on test dataset predictions

```python
np.sqrt(mean_squared_error(Ytest, y_pred))
```

## Train the model on entire training data

#### Normalize the data

```python
train = train_df.values
train.shape
```

```python
scaler = MinMaxScaler().fit(train)
train = scaler.transform(train)
```

#### Dataset preparation

```python
sales_list = train.ravel()
_, sales_df = prepare_data(sales_list)
```

```python
train_y = np.array([sales_df['target'].values]).T
sales_df.drop(['target'], inplace=True, axis=1)
train_x = sales_df.values
print("train_x: {} \ntrain_y: {}".format(train_x.shape, train_y.shape))
```

```python
Xtrain = train_x.copy()
Ytrain = train_y.copy()
```

#### Train model on entire dataset

```python
lgtrain = lgb.Dataset(Xtrain, label=Ytrain.ravel())
model = lgb.train(params, lgtrain, num_rounds, verbose_eval=1000)
```

#### Make predictions for future 90 days

```python
Ypred = []

for i in range(90):
    Xpred = sales_df[-1:]
    y_pred = model.predict(Xpred, num_iteration=model.best_iteration)
    Ypred.append(y_pred)
    
    sales_list, sales_df = prepare_data(sales_list, y_pred)
    sales_df.drop(['target'], inplace=True, axis=1)
```

#### Plot future predictions

```python
plt.figure(figsize=(15,8))
plt.plot(np.arange(0, len(Ytrain)), Ytrain, 'g', label="history")
plt.plot(np.arange(len(Ytrain), len(Ytrain) + len(Ypred)), Ypred, 'r', label="prediction")
plt.ylabel('Value')
plt.xlabel('Time Step')
plt.legend()
plt.grid()
plt.show()
```

## Create submission file

```python
final_pred = np.array(Ypred)
final_pred = scaler.inverse_transform(final_pred)
print(final_pred.shape)
```
