```python
import numpy as np
```

```python
class StepDecay:
	def __init__(self, initAlpha=0.01, leFactor=0.5, dropEvery=10, deFactor=0.9):
		# store the base initial learning rate, drop factor, and
		# epochs to drop every
		self.initAlpha = initAlpha
		self.leFactor = leFactor
		self.dropEvery = dropEvery
		self.deFactor = deFactor

	def call(self, epoch):
		# compute the learning rate for the current epoch
		exp = np.floor((1 + epoch) / self.dropEvery)
		alpha = np.floor(self.dropEvery * (0.9 ** exp))
		#alpha = self.initAlpha * (self.factor ** exp)

		# return the learning rate
		return float(alpha)
```

```python
sd = StepDecay(initAlpha=0.01, factor=0.25, dropEvery=10)
```

```python
for idx, i in enumerate(range(100)):
    x = sd.call(i)
    print("{} | {}".format(idx+1, x))
```
