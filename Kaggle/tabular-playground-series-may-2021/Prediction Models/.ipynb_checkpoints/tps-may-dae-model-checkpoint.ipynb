## Import libraries

```python
import gc
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from tensorflow.keras.utils import Sequence
from tensorflow.keras.regularizers import l2
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow_addons.optimizers import AdamW, Lookahead
from tensorflow.keras.layers import Concatenate
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import SpatialDropout1D, Add
from tensorflow.keras.layers import Dense, Input, Reshape
from tensorflow.keras.layers import Activation, Embedding
```

## Prepare data for model training

```python
train_df = pd.read_csv("../input/tabular-playground-series-may-2021/train.csv")
test_df = pd.read_csv("../input/tabular-playground-series-may-2021/test.csv")

train_df.set_index('id', inplace=True)
test_df.set_index('id', inplace=True)
train_df.drop(['target'], inplace=True, axis=1)

print("train_df: {} \ntest_df: {}".format(train_df.shape, test_df.shape))
```

```python
test_df.loc[test_df['feature_3']==25,'feature_3']=26
test_df.loc[test_df['feature_4']==36,'feature_4']=37
test_df.loc[test_df['feature_21']==31,'feature_21']=36
test_df.loc[test_df['feature_25']==24,'feature_25']=23
test_df.loc[test_df['feature_34']==26,'feature_34']=25
test_df.loc[test_df['feature_49']==21,'feature_49']=20

train_df = train_df[train_df['feature_5']!=10]

train_df = train_df[train_df['feature_6']!=26]
train_df = train_df[train_df['feature_6']!=27]

train_df = train_df[train_df['feature_7']!=30]
train_df = train_df[train_df['feature_7']!=31]

train_df = train_df[train_df['feature_9']!=17]

train_df = train_df[train_df['feature_10']!=16]

train_df = train_df[train_df['feature_11']!=12]

train_df = train_df[train_df['feature_15']!=20]

train_df = train_df[train_df['feature_16']!=18]

train_df = train_df[train_df['feature_23']!=18]
train_df = train_df[train_df['feature_23']!=19]

train_df = train_df[train_df['feature_27']!=29]

train_df = train_df[train_df['feature_28']!=23]

train_df = train_df[train_df['feature_29']!=13]

train_df = train_df[train_df['feature_33']!=24]

train_df = train_df[train_df['feature_32']!=26]
train_df = train_df[train_df['feature_32']!=27]

train_df = train_df[train_df['feature_35']!=43]
train_df = train_df[train_df['feature_35']!=-2]
train_df = train_df[train_df['feature_35']!=38]
train_df = train_df[train_df['feature_35']!=39]


train_df = train_df[train_df['feature_38']!=65]
train_df = train_df[train_df['feature_38']!=55]
train_df = train_df[train_df['feature_38']!=-8]
train_df = train_df[train_df['feature_38']!=-3]
train_df = train_df[train_df['feature_38']!=-2]
train_df = train_df[train_df['feature_38']!=63]

train_df = train_df[train_df['feature_39']!=65]
train_df = train_df[train_df['feature_39']!=66]
train_df = train_df[train_df['feature_39']!=-5]
train_df = train_df[train_df['feature_39']!=-3]
train_df = train_df[train_df['feature_39']!=-2]
train_df = train_df[train_df['feature_39']!=63]

train_df = train_df[train_df['feature_42']!=37]
train_df = train_df[train_df['feature_42']!=-2]
train_df = train_df[train_df['feature_42']!=-1]

train_df = train_df[train_df['feature_43']!=33]
train_df = train_df[train_df['feature_43']!=31]

train_df.shape, test_df.shape
```

```python
combined_df = train_df.append(test_df, sort=False, ignore_index=True)
combined_df.head()
```

```python
mini_batch_size = 128

Xtrain, Xtest = train_test_split(combined_df, test_size=0.2)
print("Xtrain: {} \nXtest: {}".format(Xtrain.shape, Xtest.shape))
```

```python
num_cols = ['feature_14','feature_19','feature_30','feature_31','feature_32',
            'feature_35','feature_38','feature_39','feature_42']
cat_cols = [col for col in Xtrain.columns if col not in num_cols]
len(num_cols), len(cat_cols)
```

## Build the model

```python
def autoencoder(data, cat_cols, num_cols, embed=256):
    
    inputs = []
    outputs = []
    for c in cat_cols:
        num_unique_values = int(data[c].nunique())
        embed_dim = int(min(np.ceil((num_unique_values)/2), 30))
        inp = Input(shape=(1,))
        out = Embedding(num_unique_values + 2, embed_dim, name=c)(inp)
        out = Reshape(target_shape=(embed_dim, ))(out)
        inputs.append(inp)
        outputs.append(out)

    x1 = Concatenate()(outputs)
    x1 = BatchNormalization()(x1)
    
    x2 = Input(shape=(num_cols, ), name='Input-continuous_features')
    x = Concatenate()([x1, x2])

    xe1 = Dense(units=384, name='Encoder-Dense-1', kernel_initializer='he_uniform', 
                kernel_regularizer=l2(0.0005))(x)
    xe1 = BatchNormalization(name='Encoder-BatchNorm-1')(xe1)
    xe1 = Activation('swish', name='Encoder-Swish-1')(xe1)

    xe2 = Dense(units=300, name='Encoder-Dense-2', kernel_initializer='he_uniform', 
                kernel_regularizer=l2(0.0005))(xe1)
    xe2 = BatchNormalization(name='Encoder-BatchNorm-2')(xe2)
    xe2 = Activation('swish', name='Encoder-Swish-2')(xe2)
    
    xe3 = Dense(units=384, name='Encoder-Dense-3', kernel_initializer='he_uniform', 
                kernel_regularizer=l2(0.0005))(xe2)
    xe3 = BatchNormalization(name='Encoder-BatchNorm-3')(xe3)
    xe3 = Activation('swish', name='Encoder-Swish-3')(xe3)
    xe3 = Add(name='Encoder-Add-3')([xe1, xe3])

    xe3 = Dense(units=embed, name='Embed-Dense', kernel_initializer='he_uniform', 
                kernel_regularizer=l2(0.0005))(xe3)
    xe3 = BatchNormalization(name='Embed-BatchNorm')(xe3)
    embed = Activation('swish', name='Embedding')(xe3)
    
    xd1 = Dense(units=196, name='Decoder-Dense-1', kernel_initializer='he_uniform', 
                kernel_regularizer=l2(0.0005))(embed)
    xd1 = BatchNormalization(name='Decoder-BatchNorm-1')(xd1)
    xd1 = Activation('swish', name='Decoder-Swish-1')(xd1)

    xd2 = Dense(units=128, name='Decoder-Dense-2', kernel_initializer='he_uniform', 
                kernel_regularizer=l2(0.0005))(xd1)
    xd2 = BatchNormalization(name='Decoder-BatchNorm-2')(xd2)
    xd2 = Activation('swish', name='Decoder-Swish-2')(xd2)
    
    xd3 = Dense(units=196, name='Decoder-Dense-3', kernel_initializer='he_uniform', 
                kernel_regularizer=l2(0.0005))(xd2)
    xd3 = BatchNormalization(name='Decoder-BatchNorm-3')(xd3)
    xd3 = Activation('swish', name='Decoder-Swish-3')(xd3)
    xd3 = Add(name='Decoder-Add-3')([xd1, xd3])

    x_output = Dense(units=data.shape[1], name='Decoder-Output', 
                     kernel_initializer='he_uniform')(xd3)

    model = Model(inputs=[inputs, x2], outputs=x_output, name='Autoencoder_Model')
    return model
```

```python
model = autoencoder(combined_df, cat_cols, len(num_cols))
model.compile(loss='mean_squared_error',
              optimizer=Lookahead(AdamW(lr=1e-2, 
                                        weight_decay=1e-5, 
                                        clipvalue=700), 
                                  sync_period=10))

early = EarlyStopping(monitor="val_loss", mode="min", 
                      restore_best_weights=True, 
                      patience=12, verbose=1)

reduce_lr = ReduceLROnPlateau(monitor="val_loss", factor=0.1, 
                              min_lr=1e-6, patience=5, 
                              verbose=1, mode='min')

chk_point = ModelCheckpoint('./DAE_model.h5', 
                            monitor='val_loss', verbose=1, 
                            save_best_only=True, mode='min')
```

```python
model.summary()
```

```python
history = model.fit(
    [[Xtrain[col] for col in cat_cols], Xtrain[num_cols]],
    Xtrain,
    batch_size=mini_batch_size,
    epochs=300, 
    verbose=1, 
    workers=5,
    callbacks=[reduce_lr, early, chk_point], 
    validation_data=([[Xtest[col] for col in cat_cols], Xtest[num_cols]],
                     Xtest)
)
```

```python
plt.figure(figsize=(15,8))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper right')
plt.show()
```
