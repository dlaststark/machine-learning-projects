## Import Packages

```python
from os import listdir
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

## Extract and store code snippets

```python
dataset = {}
for folder in listdir('Dataset'):
    for idx, file in enumerate(listdir('Dataset/' + folder)):
        print("{} | {} -> {}".format(idx, folder, file))
        f = open('Dataset/' + folder + "/" + file, "r")
        dataset[folder + "_" + str(idx)] = f.read()
        f.close()
```

```python
df = pd.DataFrame(list(dataset.items()), columns=['prog_lang', 'code_snippet'])
df['prog_lang'] = df['prog_lang'].apply(lambda x: x.split('_')[0])
print(df.shape)
df.head()
```

## Exploratory Data Analysis

```python
# lowercase everything
df['code_snippet'] = df['code_snippet'].apply(lambda x: x.lower())

# get rid of '\n' from whitespace 
df['code_snippet'] = df['code_snippet'].apply(lambda x: x.replace('\n', ' '))
df.head()
```

### Get number of code snippets in each programming language

```python
plt.figure(figsize=(15, 10))
plt.xticks(rotation=90)
sns.countplot(x="prog_lang", data=df)
```

### Understanding common words in few sample programming languages

#### 1)  Java

```python
all_words = ' '.join([text for text in df[df['prog_lang']=='java']['code_snippet']])
wordcloud = WordCloud(width=800, height=500, random_state=21, max_font_size=110).generate(all_words)
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```
